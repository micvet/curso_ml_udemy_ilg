### Trabalhando com Iris Dataset


```python
import sklearn
from sklearn.datasets import load_iris
iris = load_iris()

```

Em geral, usamos o X para as observações e o Y para target

##Observações

```python
x = iris.data
print (x)
```

##Target

```python
y= iris.target
print(y)
#os arrays devem ser representados por numeros
print (iris.target_names)
```

##Shape das observaçoes
Verficando se o numero de observações e de target estão iguais

```python
print (iris.data.shape)
```

##Shape target

```python
print(iris.target.shape)
```

Para avaliação, usaremos o modelo KNN (K nearest neighbor), que usa as amostras mais próximas para gerar uma previsão

```python
from sklearn.neighbors import KNeighborsClassifier
#o parâmetro apresentado (1) é a quantidade de neighbors que devem ser avaliadas para previsão
knn = KNeighborsClassifier(n_neighbors=1)
```

##Treinando nossa máquina com nossos dados x e y

```python
knn.fit(x,y)
```

##Fazendo previsões

```python
especie = knn.predict([[5.9,3,5.1,1.8]])
#o array que retorna indica a espécie
#podemos fazer dessa forma para retornar o nome:
print (iris.target_names[especie])
```

##Avaliação da performance do modelo

```python
#primeiro separamos os dados em dois grupos
#precisamos separar os dados em treino e teste
from sklearn.model_selection import train_test_split
#abaixo estamos separando nossos dados x e y em treino e teste. O parâmetro test_size indica a porcentagem de dados que serão utilizados para o teste
#O ideal é sempre deixar uma maior parte para o treino. Abaixo teremos 75% para treino e 25% para teste
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
#treinando a maquina com os dados de treino
knn.fit(x_train, y_train)
#para previsao utilizaremos o x_test, como se estivéssemos colocando valores para a maquina prever
previsao = knn.predict(x_test)
print (previsao)
```

```python
#agora vamos avaliar o accuracy_score das nossas previsões
from sklearn import metrics
acertos = metrics.accuracy_score(y_test, previsao)
print(acertos)
#quanto mais dados, maior é nossa precisão
#a alteração do K pode causar muita variação nos acertos
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
previsao = logreg.fit(x_train, y_train)
print (logreg.predict([[5.9,3,5.1,1.8]]))
#proba utiliza o método one vs all
print (logreg.predict_proba([[5.9,3,5.1,1.8]]))

previsao_logreg = logreg.predict(x_test)
acertos_logreg = metrics.accuracy_score(y_test, previsao_logreg)
print (acertos_logreg)
```
