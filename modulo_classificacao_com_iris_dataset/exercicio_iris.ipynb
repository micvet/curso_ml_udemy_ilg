#O exercício proposto é de alterar o K em 25 vezes e verificar número que apresenta maior taxa de acerto.

```python
import sklearn
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
k = 1
performance = {}
while k <26:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train, y_train)
  previsao = knn.predict(x_test)
  acertos = metrics.accuracy_score(y_test, previsao)
  performance[k] = round(acertos,4)
  k +=1
print (performance)
```

##Agora vamos colocar os valores em um gráfico para visualizarmos melhor os resultados

```python
import matplotlib.pyplot as plt
#para plotar o gráfico teremos que passar os valores do dicionário para listas. Faremos lista de chaves e valores
plt.bar(list(performance.keys()), list(performance.values()))
plt.xlabel('Valores K')
plt.ylabel('Performance')
plt.show
```
